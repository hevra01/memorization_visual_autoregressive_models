```python
import os
import sys
PROJECT_ROOT = os.path.abspath("../..")
sys.path.insert(0, PROJECT_ROOT)

print("Project root:", PROJECT_ROOT)

import torch

```

```python
# nested within this directory are folders for each augmentation run (0 to 9)
activation_base_dir = "/scratch/inf0/user/hpetekka/var_mem/output_activations"

# transformer block count
num_blocks = 16

keys = [
    "fc1", "fc1_act",
    "fc2", "attn_proj",
    "k", "v", "q"
]

# dictionary to hold activations
data = {}
for block_i in range(num_blocks):
    data[block_i] = {}
    for key in keys:
        data[block_i][key] = None

def create_activation_dict(num_blocks, keys):
    activation_dict = {}
    for block_i in range(num_blocks):
        activation_dict[block_i] = {}
        for key in keys:
            activation_dict[block_i][key] = None
    return activation_dict
```

```python
dict_act = create_activation_dict(num_blocks, keys[:2])
```

```python
aug_0 = os.path.join(activation_base_dir, "0")
aug_0_activation_dict = create_activation_dict(num_blocks, keys)

for block_idx in range(1):
    for key in keys:
        fc1_block_0 = torch.load(os.path.join(aug_0, key, f"block_{block_idx}_combined.pt"))
        aug_0_activation_dict[block_idx][key] = fc1_block_0
```

```python
print(aug_0_activation_dict.keys())
print(aug_0_activation_dict[0].keys())
print(aug_0_activation_dict[0]["fc1"].shape)
```

```python
# print mean and std of activations for block 0, fc1, augmentation 0
for idx, key in enumerate(aug_0_activation_dict[0]):
    #print(f"Activation shape for block 0, key {key}:", aug_0_activation_dict[0][key].shape)
    fc1_block_0_activations = aug_0_activation_dict[0][key]
    print(f"Mean of {key} block 0 activations:", f"{torch.mean(fc1_block_0_activations).item():.02f}")
    print(f"Std of {key} block 0 activations:", f"{torch.std(fc1_block_0_activations).item():.02f}\n\n")
```
